```python
from xgboost import XGBClassifier
```

```python
tr_data=pd.read_csv('Training.csv')
```

```python
test_data=pd.read_csv('Testing.csv')
```

```python
tr_data.head()
```

```python
X_train=tr_data.iloc[:, :-2].values
X_train.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
y_train=tr_data.iloc[:, 132].values
y_train=lbl.fit_transform(y_train)
y_train.shape
```

```python
xgboost=XGBClassifier()
```

```python
xgboost.fit(X_train,y_train)
```

```python
X_test=test_data.drop(columns='prognosis',axis=1)
X_test.shape
```

```python
y_test=test_data['prognosis']
y_test=lbl.fit_transform(y_test)
y_test.shape
```

```python
y_pred=xgboost.predict(X_test)
y_pred.shape
```

```python
pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
}).head(10)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
